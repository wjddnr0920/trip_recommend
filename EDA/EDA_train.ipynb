- train_clean_custom.csv는 train_custom.csv와는 다르게 supercategory 열이 없음
- 위키피디아 크롤링을 했음에도 description이 없는 landmark는 제거

```python
train_clean_custom_df = pd.read_csv('/home/workspace/data/GLDv2/train/train_clean_custom.csv')
```

```python
train_clean_custom_df.drop('supercategory', axis=1, inplace=True)
```

```python
train_clean_custom_df.info()
```

```python
# ① 보조 DF에서 각 landmark_id별 대표값(중복 시 첫 행만) 준비
lookup = (
    train_clean_missing_custom_df[["landmark_id", "description", "language"]]
)

# ② dict 로 변환
desc_map = dict(zip(lookup["landmark_id"], lookup["description"]))
lang_map = dict(zip(lookup["landmark_id"], lookup["language"]))

# ③ NaN 인 곳만 채우기
train_clean_custom_df["description"] = train_clean_custom_df["description"].fillna(
    train_clean_custom_df["landmark_id"].map(desc_map)
)
train_clean_custom_df["language"] = train_clean_custom_df["language"].fillna(
    train_clean_custom_df["landmark_id"].map(lang_map)
)
```

```python
train_clean_custom_df.dropna(inplace=True)
```

```python
train_clean_custom_df.info()
```
